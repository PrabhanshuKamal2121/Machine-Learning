```python
df[['is_Internship_Experience','is_Placement']]=df[['is_Internship_Experience','is_Placement']].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler

col = ['IQ', 'Prev_Sem_Result', 'CGPA', 'Academic_Performance', 'Extra_Curricular_Score',
       'Communication_Skills', 'Projects_Completed']

scaler = StandardScaler() 
df[col]=scaler.fit_transform(df[col]) 
df.head()
```

```python
from scipy.stats import pearsonr

selected_features = ['IQ', 'Prev_Sem_Result', 'CGPA', 'Academic_Performance', 
                     'Extra_Curricular_Score', 'Communication_Skills', 'Projects_Completed']

# Calculate correlation coefficients
correlations = {
    feature: pearsonr(df[feature], df['is_Placement'])[0]
    for feature in selected_features
}

# Convert to DataFrame
correlations_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson_Correlation'])

# Sort by correlation value (descending)
correlations_df = correlations_df.sort_values(by='Pearson_Correlation', ascending=False)
correlations_df
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05
chi2_results = {}

contigency = pd.crosstab(df['is_Internship_Experience'],df['is_Placement'])
chi2_stat,p_val,_,_ = chi2_contingency(contigency)
decision = 'Reject null(keep feature)' if p_Val<alpha else 'Accept Null(Drop Feature)'
chi2_results['is_Internship_Experience'] = {
        'chi2_statistic': chi2_stat,
        'p_value':p_val,
        'Decision': decision
}

chi2_results_df= pd.DataFrame(chi2_results).T
chi2_results_df = chi2_results_df.sort_values(by='p_value')
chi2_results_df
```
