```python
X = df.drop(columns='price',axis=1)
Y =df['price']
```

```python
X_one_encode = pd.get_dummies(data=X,columns=['model','transmission','fuelType'],drop_first=True,dtype='int')
X_one_encode
```

```python
from sklearn.preprocessing import LabelEncoder
columns = ['model','transmission','fuelType']
Xlabel = X.copy() 
label_encoders = {}

for col in columns: 
    le = LabelEncoder() 
    Xlabel[col]= le.fit_transform(Xlabel[col].astype(str))
    label_encoders[col] = le 

Xlabel
```

```python
X_one_encode.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
tobeScaled = ['year', 'mileage', 'tax', 'mpg','engineSize']
X_one_encode[tobeScaled] = scaler.fit_transform(X_one_encode[tobeScaled])
X_one_encode
```

```python
Xlabel.columns
```

```python
label_scaled = ['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']
Xlabel[label_scaled] = scaler.fit_transform(Xlabel[label_scaled])
Xlabel
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encode, Y, test_size=0.20, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
y_pred
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n=X_test.shape[0]
p=X_test.shape[1]
adjusted_r2 = 1 - ((1-r2)*(n-1))/(n-p-1)
print('Adjusted R2 Score:',adjusted_r2)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(Xlabel, Y, test_size=0.20, random_state=42)
```

```python
model = LinearRegression()
model.fit(X1_train,y1_train)
```

```python
y1_pred = model.predict(X1_test)
r2 = r2_score(y1_test,y1_pred)
r2
```

```python
n=X1_test.shape[0]
p=X1_test.shape[1]
adjusted1_r2 = 1 - ((1-r2)*(n-1))/(n-p-1)
print('Adjusted R2 Score:',adjusted1_r2)
```
