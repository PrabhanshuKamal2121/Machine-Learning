```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh')
```

```python
df.head()
```

```python
def wicket_extraction(txt):
    data=''
    for i in txt:
        for j in range(0,1):
            data=data+txt[j]
        return data

df['new_wickets']=df['best_bowling_figure'].apply(wicket_extraction)
df['new_wickets'] = df['new_wickets'].astype(int)
```

```python
df.head()
```

```python
bowling_best = df.groupby('best_bowling')['new_wickets'].sum().sort_values(ascending=False).head(10)
bowling_best
```

```python
bowling_best.plot(kind='barh')
```

```python
df.head()
```
