```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y,color="black")
plt.title("hello")
plt.xlabel("chacha")
plt.ylabel("chachi")
```

```python
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
```

```python
fig = plt.figure(figsize=(12,8))
axis1=fig.add_axes([0.1,0.1,0.4,0.8])
axis1.plot(x,y,linestyle='-.',marker='^')
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```
