```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import make_blobs 
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')
```

```python
X,y = make_blobs(n_samples=500,n_features=5,centers=3,cluster_std=1.5,random_state=42)
```

```python
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

#### pehle 5 the columns/features(n_features=5) abd bas 2 bache hai because of n_components=2

```python
X_pca
```

```python
df_pca = pd.DataFrame(X_pca, columns=['PC1','PC2'])
df_pca['label']=y
```

```python
sns.scatterplot(data=df_pca,x='PC1', y='PC2',hue = 'label',palette='Set2')
```
