from geopy.geocoders import Nominatim
from geopy.distance import geodesic

def distance_calculator(country):
    geolocator = Nominatim(user_agent="distance_calc")
    loc1 = geolocator.geocode("India")
    loc2 = geolocator.geocode(country)
    distance_km = geodesic((loc1.latitude, loc1.longitude), (loc2.latitude, loc2.longitude)).kilometers
    return int(distance_km)

df_test1['DISTANCE_TO_COVER']=df_test1['FOREIGN_COUNTRY'].apply(distance_calculator)
