```python
categorical_columnms=['Family_History_of_Diabetes','Ethnicity_Asian', 'Ethnicity_Black', 'Ethnicity_Hispanic','Ethnicity_White', 'Physical_Activity_Level_High', 'Physical_Activity_Level_Low', 'Physical_Activity_Level_Moderate','Smoking_Status_Current', 'Smoking_Status_Former','Smoking_Status_Never', 'BMI_category_underweight','BMI_category_normal', 'BMI_category_overweight', 'BMI_category_obese']
```

```python
numerical_columns = ['Age', 'Sex', 'BMI', 'Waist_Circumference', 'Fasting_Blood_Glucose', 'Blood_Pressure_Systolic', 'Blood_Pressure_Diastolic','Cholesterol_Total', 'Cholesterol_HDL', 'Cholesterol_LDL', 'GGT','Serum_Urate', 'Dietary_Intake_Calories']
```

```python
df['HbA1c'] = pd.cut(
    df['HbA1c'],
    bins = [0, 5.7,float('inf')], 
    labels = [0,1],
)
```

```python
from scipy.stats import pearsonr

correlations = {
    feature: pearsonr(df[feature], df['HbA1c'])[0]  # [0] gives the correlation coefficient
    for feature in numerical_columns
}

correlations = pd.DataFrame(
    list(correlations.items()),
    columns=['Feature', 'Pearson_Correlation']
)

correlations = correlations.sort_values(by='Pearson_Correlation', ascending=False)
correlations
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05


##qcut- Divides charges into 4 bins (quartiles) with ~25% of data in each.
##cut- Divides the range of charges into 4 equal-width intervals.
chi2_results = {}

for col in categorical_columnms: 
    contigency = pd.crosstab(df[col],df['HbA1c']) 
    chi2_stat, p_val,_,_ = chi2_contingency(contigency)
    decision = 'Reject Null(keep feature)' if p_val<alpha else 'Accept Null(Drop Feature)'
    chi2_results[col]={
        'chi2_statistic': chi2_stat,
        'p_value':p_val,
        'Decision': decision
    }
chi2_results_df = pd.DataFrame(chi2_results).T
chi2_results_df = chi2_results_df.sort_values(by='p_value')
chi2_results_df
```
