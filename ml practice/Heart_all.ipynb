```python
categoric_columns = ['Sex','FastingBS','ExerciseAngina','ChestPainType_ASY',
       'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA',
       'RestingECG_LVH', 'RestingECG_Normal', 'RestingECG_ST', 'ST_Slope_Down',
       'ST_Slope_Flat', 'ST_Slope_Up', 'Cholesterol_types_Low',
       'Cholesterol_types_normal', 'Cholesterol_types_high']
```

```python
numeric_columns = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR','Oldpeak']
```

```python
from scipy.stats import pearsonr
```

```python
correlations = {
    feature: pearsonr(df[feature],df['HeartDisease'])[0]
    for feature in numeric_columns
}

correlations_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson_Correlation'])

# Sort (optional)
correlations_df = correlations_df.sort_values(by='Pearson_Correlation', ascending=False)

correlations_df
```

```python
# Cholesterol	-0.012340
```

```python
from scipy.stats import chi2_contingency

chi2_results = {} 
alpha = 0.02


for col in categoric_columns: 
    contingency = pd.crosstab(df[col],df['HeartDisease']) 
    chi2_stats,p_value,_,_=chi2_contingency(contingency)
    decision = 'AcceptNUll(RejectFeature)' if p_value>alpha else 'RejectNull(AcceptFeature)'
    chi2_results[col]={
        'chi2_value':chi2_stats,
        'p_value' :p_value, 
        'decision' : decision
    }

chi2_df = pd.DataFrame(chi2_results).T
```
