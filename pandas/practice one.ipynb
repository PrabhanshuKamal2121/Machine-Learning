```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i ==')':
            check=False
            return data
        if check==True:
            data=data+i
        if i=='(':
            check=True
df["Episodes"]=df["Title"].apply(extract_episodes)
```

```python
def Time_stamps(txt):
    start=False
    data =""
    for i in range(len(txt)):
        if txt[i]==')':
            start=True
            for j in range(i+1,i+20):
                data=data+txt[j]
            return data

df["TimeStamp"]=df["Title"].apply(Time_stamps)
```

```python
def month_extraction(txt):
    data=""
    for i in range(len(txt)):
        for j in range(i,i+3):
            data =data +txt[j]
        for k in range(i+8,i+14):
            data=data+txt[k]
        return data

df["Month"]=df["TimeStamp"].apply(month_extraction)
df
```

```python
df[['Start', 'End']] = df['TimeStamp'].str.split(' - ', expand=True)

# Step 2: Convert both start and end into datetime format
# This assumes day=1 for both dates
df['Start'] = pd.to_datetime(df['Start'], format='%b %Y', errors='coerce')
df['End'] = pd.to_datetime(df['End'], format='%b %Y', errors='coerce')

# Step 3: Compute the duration in days
df['Duration_Days'] = (df['End'] - df['Start']).dt.days

df
```
